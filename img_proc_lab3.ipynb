```python
from show_img import show_img
import numpy as np
import cv2
```

## 1. Перевод в грейтскейл с диапазоном

```python
img = cv2.imread("bibizyana.jpg")
show_img(img)
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_img(img_grayscale)
range_min, range_max = 100, 150

img_grayscale_range = img_grayscale / 255 * (range_max - range_min) + range_min
img_grayscale_range = img_grayscale_range.astype(np.uint8)
show_img(img_grayscale_range)
```

## 2. Обратное преобразование

```python
img_grayscale_reverse = (img_grayscale_range - range_min) / (range_max - range_min) * 255
img_grayscale_reverse = img_grayscale_reverse.astype(np.uint8)
show_img(img_grayscale_reverse)
```

## 3. Контрастироование

```python
x_min = np.min(img)
x_max = np.max(img)
y_min = 50
y_max = 220

img_contrast = (img - x_min) / (x_max - x_min) * (y_max - y_min) + y_min
img_contrast = img_contrast.astype(np.uint8)

#show_img(img)
show_img(img_contrast)
```

## 4. Соляризация

```python
thres = 20
img_sol = np.where(img > thres, 255 - img, img)
show_img(img_sol)
thres = 120
img_sol = np.where(img > thres, 255 - img, img)
show_img(img_sol)
thres = 220
img_sol = np.where(img > thres, 255 - img, img)
show_img(img_sol)

thres = 120
img_sol = np.where(img > thres, 255 - img, img)
show_img(img_sol, f"best with thres {thres}")
```

## 5. Размытие - нечитаемый текст

```python
value = 9
img_blur = cv2.GaussianBlur(img, (value * 2 + 1, value * 2 + 1), 0)
show_img(img_blur)
```

## 6. Двусторониий фильтр

```python
d = 25 # область вокруг пикселя
sigmaColor = 100 # разница цветов пикселей
sigmaSpace = 100 # вес влияния дальних пикселей на центральный 
img_bil = img.copy()
img_bil = cv2.bilateralFilter(img, d=d, sigmaColor=sigmaColor, sigmaSpace=sigmaSpace)
show_img(img_bil)
```

## 7. Фильтры границ

### Собель

```python
ksize = 1
img_sobel_x = cv2.Sobel(img_grayscale, cv2.CV_64F, 1, 0, ksize=ksize*2+1)
img_sobel_y = cv2.Sobel(img_grayscale, cv2.CV_64F, 0, 1, ksize=ksize*2+1)
show_img(img_sobel_x, "by x")
show_img(img_sobel_y, "by y")
```

### Лапласа

```python
img_lap = cv2.Laplacian(img_grayscale, cv2.CV_64F)
show_img(img_lap)
```

### Канни

```python
img_canny = cv2.Canny(img, 20, 80)
show_img(img_canny)
```
