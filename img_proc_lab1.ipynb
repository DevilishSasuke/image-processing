```python
def create_list(n: int) -> int:
  if n < 1:
    return 0
  np.random.seed(314)
  numbers = np.random.uniform(0, 100, n)
  numbers = np.round(numbers, 2)
  numbers = np.where(numbers < 10, numbers + 10, numbers)
  numbers = np.where(numbers > 90, numbers - 10, numbers)

  return int(np.min(numbers) + np.max(numbers) + np.mean(numbers))

create_list(10)
```

```python
def matrix_prosessing(m: int, n: int) -> int:
  matrix = np.zeros((m, n), dtype=int)

  max_depth = ceil(min(m, n) / 2)
  max_depth = 0 if m < 2 or n < 2 else max_depth
  for i in range(max_depth, -1, -1):
    matrix[0 + i, :] = i + 1
    matrix[-(1 + i), :] = i + 1
    matrix[:, 0 + i] = i + 1
    matrix[:, -(1 + i)] = i + 1

  return int(np.sum(matrix))

matrix_prosessing(4,5)
```

```python
def temperature_prosessing(n: int) -> int:
  np.random.seed(314)
  temp1, temp2 = np.random.uniform(15, 33, 30), np.random.uniform(15, 33, 30)
  temp_gen = np.hstack((temp1[temp1 > n], temp2[temp2 > n]))
  return int(temp_gen.size + np.mean(temp_gen) + np.min(temp_gen) + np.max(temp_gen) + np.std(temp_gen))


temperature_prosessing(24)
```
